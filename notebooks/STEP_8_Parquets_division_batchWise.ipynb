```python
import pandas as pd
import os

batch_size_map = {}

# First batch range (batch size = 4)
for i in range(0, 47 * 4, 4):
    batch_size_map[i] = 4

# Second batch range (batch size = 1)
for i in range(192, 515):
    batch_size_map[i] = 1

# os.makedirs("parquet_batches", exist_ok=True)
#saves the new batch wise parquet with showing the starting row of the batch in name of the file
for batch_start, bsize in batch_size_map.items():
    batch_df = df.iloc[batch_start:batch_start + bsize]
    batch_df.to_parquet(f"{path}/risk_matrix_{batch_start}.parquet", index=False)
```
